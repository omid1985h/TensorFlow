```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Step 1: Define function to create training data
def create_training_data():
    # Example data for features (n_bedrooms) and targets (price_in_hundreds_of_tousands)
    n_bedrooms = np.array([1, 2, 3, 4, 5, 6], dtype=np.float64)  # 6 features (bedrooms)
    price_in_hundreds_of_tousands = np.array([1.5, 2.0, 2.5, 3.0, 3.5, 4.0], dtype=np.float64)  # 6 target values (price in hundreds of thousands)
    return n_bedrooms, price_in_hundreds_of_tousands

# Step 2: Define function to build and compile the model
def define_and_compile_model():
    model = Sequential([
        tf.keras.Input(shape=(1,)),  # input layer with one feature (n_bedrooms)
        Dense(units=1)  # output layer with one unit (price prediction)
    ])
    
    model.compile(optimizer=SGD(), loss='mean_squared_error')  # Use SGD optimizer and MSE loss
    return model

# Step 3: Main function to execute the workflow
def main():
    # Create training data
    features, targets = create_training_data()

    # Output the shapes of features and targets
    print(f"Features have shape: {features.shape}")
    print(f"Targets have shape: {targets.shape}")

    # Reshape the features for the model (TensorFlow expects 2D input, not 1D)
    features = features.reshape(-1, 1)  # Reshape to (6, 1)

    # Define and compile the model
    model = define_and_compile_model()

    # Step 4: Train the model
    model.fit(features, targets, epochs=100)

    # Step 5: Make predictions (optional, just to test)
    predictions = model.predict(features)
    print(f"Predictions: {predictions.flatten()}")

# Execute the script
if __name__ == "__main__":
    main()
```

## Fashion MINST



* When discussing the Fashion-MNIST dataset in the context of neural networks, particularly Convolutional Neural Networks (CNNs), it's important to understand the structure of the dataset and how that affects the design of the network architecture, especially the input layer, hidden layers, and output layer. Let's break down your questions:

#### 1. Why is the input size 28x28?
* The Fashion-MNIST dataset consists of grayscale images, where each image is 28 pixels by 28 pixels. This is the input size because each image is a 28x28 matrix of pixel values. The dataset contains images of 10 different categories of fashion items (like T-shirts, trousers, etc.), and each image has been preprocessed to a fixed size of 28x28 pixels. So, the neural network is expected to take these 28x28 pixel matrices as input.

* For CNNs, the input is typically a 3D array of shape (height, width, channels). For Fashion-MNIST:
height = 28
width = 28
channels = 1, since the images are grayscale.
Hence, the input shape is (28, 28, 1) for a single image.

#### 2. Why does the end layer have 10 units?
* The Fashion-MNIST dataset consists of 10 different classes of clothing items, like:
T-shirt/top
Trouser
Pullover
Dress
Coat
Sandal
Shirt
Sneaker
Bag
Ankle boot
* Since this is a classification problem with 10 classes, the output layer needs 10 units, one for each class. Each unit in the final layer represents the probability that the input image belongs to that particular class. The network will output a vector of 10 probabilities (values between 0 and 1), where the sum of all probabilities is 1. The class with the highest probability is the predicted class.

* In summary, the output layer has 10 neurons because there are 10 possible classes to predict.

#### 3. Why do we use multiple layers in between? (Hidden layers)
* The hidden layers are responsible for learning hierarchical features of the input data. Here's a breakdown of the layers you would typically find in a CNN architecture for Fashion-MNIST:
### Convolutional Layers
* These layers apply convolution operations to the input, detecting various features like edges, corners, textures, etc. In a deep CNN, you often stack multiple convolutional layers to progressively extract more complex features (e.g., from edges to shapes to more abstract features).
#### Activation Functions (e.g., ReLU)
* After each convolutional layer, you apply an activation function like ReLU (Rectified Linear Unit), which introduces non-linearity to the network. Without these activations, the network would behave like a linear function, no matter how many layers you stack.

#### Pooling Layers
* Max pooling is commonly used to reduce the spatial dimensions (height and width) of the feature maps produced by the convolutional layers. Pooling layers help reduce computation and prevent overfitting by summarizing the presence of features in smaller regions.

#### Fully Connected (Dense) Layers
* After several convolutional and pooling layers, the output is usually flattened into a 1D vector and passed through one or more fully connected (dense) layers. These layers combine the learned features and make the final prediction based on them.
* The final fully connected layer before the output layer typically has 128, 256, or more units. These units capture complex combinations of features to help make a final decision about the class of the image.

#### Why the network uses the specific architecture (e.g., number of layers)?
* The architecture of a CNN is often determined experimentally, but here are some general principles that guide its design:
* Depth of the network: The number of layers determines the depth of the model. Deeper networks can learn more complex representations, but they also require more data to train effectively. For Fashion-MNIST, a relatively simple architecture (with fewer layers) often works well because the dataset is not as complex as more challenging datasets like ImageNet.
* Convolutional layers: The first few layers focus on low-level features (edges, textures), and as you go deeper, the network can capture higher-level abstract features (e.g., shapes, patterns). Adding more convolutional layers can improve feature extraction, but there's a tradeoff with computational cost and potential overfitting.
* Fully connected layers: After extracting features through convolutions, the fully connected layers help the network make sense of these features and classify the image correctly. The number of neurons in these layers varies based on the complexity of the problem and the size of the network.
* Output layer (10 neurons): Since Fashion-MNIST has 10 classes, the final output layer must have 10 neurons. These neurons are often connected to a softmax activation function, which outputs a probability distribution over the 10 classes.

* Example Architecture (for Fashion-MNIST):
* A simple CNN architecture for Fashion-MNIST might look like this:
* Input layer: Size 28x28x1 (grayscale image).
* Convolutional layer: 32 filters of size 3x3, activation: ReLU.
* Max-pooling layer: Pool size 2x2.
* Convolutional layer: 64 filters of size 3x3, activation: ReLU.
* Max-pooling layer: Pool size 2x2.
* Flatten layer: Flatten the output into a 1D vector.
* Fully connected layer (Dense): 128 neurons, activation: ReLU.
* Output layer: 10 neurons (one for each class), activation: Softmax.

Conclusion:
* The input size of 28x28 comes from the fact that the Fashion-MNIST images are 28x28 pixels in grayscale.
* The output layer has 10 neurons because the classification task has 10 classes.
* The intermediate layers (convolutional, pooling, fully connected) serve the purpose of feature extraction and classification, progressively learning and combining low-level features into high-level patterns for accurate predictions.

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

# 1. Load the Fashion-MNIST dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()

# 2. Normalize the images to values between 0 and 1 (pixel values originally between 0-255)
x_train, x_test = x_train / 255.0, x_test / 255.0

# 3. Flatten the images to a 1D vector (28x28 pixels becomes 784 features)
x_train = x_train.reshape(-1, 28 * 28)
x_test = x_test.reshape(-1, 28 * 28)

# 4. Build a simple fully connected model
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28 * 28,)),  # Flatten the input
    tf.keras.layers.Dense(128, activation='relu'),    # 128 neurons in the hidden layer
    tf.keras.layers.Dense(10, activation='softmax')   # 10 output neurons for classification
])

# 5. Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# 6. Train the model
history = model.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))

# 7. Evaluate the model
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print(f"Test accuracy: {test_acc:.4f}")

# 8. Plot training results (accuracy and loss)
plt.figure(figsize=(12, 5))

# Plot accuracy
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training accuracy')
plt.plot(history.history['val_accuracy'], label='Validation accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

# Plot loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training loss')
plt.plot(history.history['val_loss'], label='Validation loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')

plt.tight_layout()
plt.show()

# 9. Visualize some input images and their predicted labels
fig, axes = plt.subplots(2, 5, figsize=(10, 5))
axes = axes.ravel()

# Predict on the test data
predictions = model.predict(x_test[:10])

# Plot the first 10 test images with their predicted labels
for i in np.arange(0, 10):
    axes[i].imshow(x_test[i].reshape(28, 28), cmap='gray')
    axes[i].set_title(f"Pred: {np.argmax(predictions[i])}, True: {y_test[i]}")
    axes[i].axis('off')

plt.tight_layout()
plt.show()
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

# 1. Load the Fashion-MNIST dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()

# 2. Normalize the images to values between 0 and 1 (pixel values originally between 0-255)
x_train, x_test = x_train / 255.0, x_test / 255.0

# 3. Reshape the images (28x28 pixels becomes 784 features)
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)

# 4. Build a simple CNN model
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')  # 10 classes for Fashion-MNIST
])

# 5. Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# 6. Train the model
history = model.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))

# 7. Evaluate the model
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print(f"Test accuracy: {test_acc:.4f}")

# 8. Plot training results (accuracy and loss)
plt.figure(figsize=(12, 5))

# Plot accuracy
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training accuracy')
plt.plot(history.history['val_accuracy'], label='Validation accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

# Plot loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training loss')
plt.plot(history.history['val_loss'], label='Validation loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')

plt.tight_layout()
plt.show()

# 9. Visualize some input images and their predicted labels
fig, axes = plt.subplots(2, 5, figsize=(10, 5))
axes = axes.ravel()

# Predict on the test data
predictions = model.predict(x_test[:10])

# Plot the first 10 test images with their predicted labels
for i in np.arange(0, 10):
    axes[i].imshow(x_test[i].reshape(28, 28), cmap='gray')
    axes[i].set_title(f"Pred: {np.argmax(predictions[i])}, True: {y_test[i]}")
    axes[i].axis('off')

plt.tight_layout()
plt.show()
```

* Normalization: Pixel values are between 0 and 255, but neural networks typically perform better when inputs are scaled between 0 and 1. We normalize the data by dividing by 255.
* Reshaping Data: The images in the dataset are 28x28 grayscale, so we reshape the images to include a single channel (28x28x1) since CNNs expect an additional color channel dimension.
* Model Definition: The model uses a simple CNN architecture: Conv2D layers are used to extract features from the images. MaxPooling2D layers reduce the spatial dimensions of the feature maps. Flatten converts the 2D feature maps into a 1D vector. Dense layers are used for classification, and the final layer has 10 neurons (one for each class in the Fashion-MNIST dataset) with a softmax activation function for multi-class classification.
* Compilation: The model is compiled using the Adam optimizer, sparse categorical cross-entropy loss (since the labels are integers), and accuracy as a metric.
* Training: The model is trained for 5 epochs using the training data (x_train, y_train), and the validation data (x_test, y_test) is used to track validation accuracy and loss.

### Add Callback

* Callbacks are designed to be passed as a list to the callbacks argument in the .fit() method, and they allow you to run specific actions during training (like logging, saving the model, early stopping, etc.).

* The on_epoch_end method is a callback method in Keras that is executed at the end of every epoch during training. You can use this method to access the current state of training, such as the loss, accuracy, or any other metrics, and perform specific actions or logging at that point in the training process.

```python
# Define custom callback to stop training when loss < 0.4
class StopLossCallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs=None):
        loss = logs.get('loss')  # Get the loss for this epoch
        if loss < 0.4:
            print(f"\nEpoch {epoch + 1}: Loss has reached {loss:.4f}. Stopping training.")
            self.model.stop_training = True  # Stop training

# Instantiate the callback
stop_loss_callback = StopLossCallback()
```
